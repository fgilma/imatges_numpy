#### Exercici 4
Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
a=np.array([[1,2],[3,4]])
print('Dimensions d\'a:',a.shape)
print('a:',a)

#a,b mateixes dimensions 2x2
b=np.array([[8,9],[4,5]])
print('\nDimensions de b:',b.shape)
print('b:',b)

print('\nSuma a+b:')
print(a+b)
print('Al tenir mateixes dimensions a i b es pot realitzar l\'operació sense problemes')

# c dimensions 1x1
c=np.array([2])
print('\nDimensions de c:',c.shape) 
print('c:',c)

print('\nSuma a+c:')
print(a+c)

print('c és un escalar, es pot fer broadcasting i el seu valor es suma a tots elements d\'a')

# d dimensions 1x2
d=np.array([2,10])
print('\nDimensions de d:',d.shape) 
print('d:',d)
print('d és un vector de dos elements, es pot fer broadcasting i el seu valor es suma a tots elements d\'a')

print('\nSuma a+d:')
print(a+d)

# e dimensions 1x3
e=np.array([2,3,4])
print('\nDimensions de e',e.shape)
print('e',e)
print('e és un vector de tres elements, no coincideix amb les dimensions de la matriu a per tant no es pot fer broadcasting i dóna error')
a+e
```
